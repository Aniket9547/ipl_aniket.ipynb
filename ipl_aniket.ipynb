```python


# Assuming 'data' is your DataFrame
team1 = data['team1']

# Count the occurrences of each team
team_counts = team1.value_counts()

# Plot the team counts as a bar chart
plt.bar(team_counts.index, team_counts.values)
plt.xlabel('Team')
plt.ylabel('Number of Occurrences')
plt.title('Frequency of Team1')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability if needed
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python

# Assuming 'data' is your DataFrame
team1 = data['team2']

# Count the occurrences of each team
team_counts = team1.value_counts()

# Plot the team counts as a bar chart
plt.bar(team_counts.index, team_counts.values)
plt.xlabel('Team')
plt.ylabel('Number of Occurrences')
plt.title('Frequency of Team2')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability if needed
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
sns.barplot(x='toss_winner', y='toss_decision', data=data)
plt.xlabel('Toss Winner')
plt.ylabel('Toss Decision')
plt.title('Toss Decisions by Toss Winners')
plt.xticks(rotation=45)
plt.show()
```
